```python
#Libraries
import pandas as pd
```

```python
#load data from csv 
data = pd.read_csv ('diabetic_data1.csv')
```

```python
#Show data frame details 
data.describe().transpose()
```

```python
#Drop encounter_ID 
```

```python
#clean whole data frame from ?s
data=data.replace ('?','')
data=data.replace (' ','')
```

```python
#Drop all nominal values and then run correlation statistics
#replace nominal to numeric to be able to run correlations 
data = data.replace('No',0)
data = data.replace('NO',0)
data = data.replace('Yes',1)
```

```python
#convert objects to integer
data["diabetesMed"]=data["diabetesMed"].astype(int)

  
```
